```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
x = np.random.randint(50,size=500)
y = np.random.randint(50,size=500)
df = pd.DataFrame({'x': x,'y': y})
df.head()
```

```python
px.scatter(df,x=df['x'],y=df['y'],color='x',size='y',hover_name='x',title='Customized Plot',opacity=0.7)
```
