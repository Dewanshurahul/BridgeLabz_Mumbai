```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ((np.random.randn(50))
y = ((np.random.randn(50))
```

```python
df = pd.DataFrame({'x': x,'y': y})
```

```python
plt.scatter(df['x'],df['y'],s=70,facecolor='none',edgecolors='blue')
plt.ylim(-1,11)
plt.xlim(-1,11)
plt.show()
```
