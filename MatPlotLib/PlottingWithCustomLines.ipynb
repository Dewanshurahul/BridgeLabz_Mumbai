```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2,15,11,24,23])
```

```python
plt.plot(x,x**2,label='X-Squared',lw=3,color='green')
plt.plot(x,x**3,label='X-Cubed',lw=8,color='purple')
plt.title('Squared And Cubed Graph')
plt.legend()
```
