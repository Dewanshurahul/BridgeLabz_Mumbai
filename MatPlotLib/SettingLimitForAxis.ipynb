```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = np.array([2,15,11,24,23])
```

```python
plt.plot(x,x**2,label='X-Squared',color='green',marker="*",markersize=10)
plt.plot(x,x**3,label='X-Cubed',color='purple',marker="o")
plt.title('Squared And Cubed Graph')
plt.legend()
```
